```python
# Data Loading Code Hidden Here
import pandas as pd
# Load data
melbourne_file_path = '/content/gdrive/MyDrive/Colab_Notebooks/Melbourne_housing.csv'
melbourne_data = pd.read_csv(melbourne_file_path, low_memory=False)
```

```python
# Filter rows with missing price values
filtered_melbourne_data = melbourne_data.dropna()
# Choose target and features
```

```python
y = filtered_melbourne_data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Latitude', 'Longtitude']
X = filtered_melbourne_data[melbourne_features]
```

```python
from sklearn.tree import DecisionTreeRegressor
# Define model
# 10, 50 ,100, 500, 1000
melbourne_model = DecisionTreeRegressor()

# Fit model
melbourne_model.fit(X, y)
```

```python
yp=melbourne_model.predict(X)
```

Once we have a model, here is how we calculate the mean absolute error:

```python
yp = melbourne_model.predict(X)
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, yp)
```

```python
y.mean()
```

```python
(mean_absolute_error(y, yp)/y.mean())*100
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(n_estimators=1000,max_leaf_nodes=500,random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

```python
(mean_absolute_error(val_y, melb_preds)/y.mean())*100
```
